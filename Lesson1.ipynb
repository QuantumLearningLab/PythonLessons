Next, we will introduce "variables" with are names that are given items that corresponds to how it is stored in the computer's memory. Though not required, we are going to use camelCase to name variables.
* Variables start with a lowercase character
* Multiword variable names use a Capital letter to delininate words

```python
year = 2025
myName = "Brian A Rashap"
myBirthMonth = "April"
myBirthYear = 1968

print("Hi, I am",myName,". On my birthday, in the month of",myBirthMonth,"I was/will be", year-myBirthYear)
```

As you can see, we can use the variables in mathematical expressions or print statements. Though, the print statements become clunky. We can simplify the code a bit by using formatted print statements as in the below. Note the f inside the print's parentheses

```python
print(f"Hi, I am {myName}. On my birthday, in the month of {myBirthMonth}, I was/will be {year-myBirthYear}")
```
