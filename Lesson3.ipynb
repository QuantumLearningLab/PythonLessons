Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

Let's create a simple plot of this data:

v | current

1 | 2.17

2 | 4.65

3 | 6.55

4 | 8.91

5 | 10.7

```python
voltage = np.array([0,1,2,3,4,5])
current = np.array([0,2.17,4.65,6.55,8.91,10.7])
plt.plot(voltage,current,'g*')
```

Roll dice 36 times and record the sums

```python
numRolls = 36
```

```python
rolls = []
for i in range(0,numRolls):
  roll = random.randrange(1,7) + random.randrange(1,7)
  rolls.append(roll)
rollArray = np.array(rolls)
print(rollArray)


```

Create a histogram


```python
plt.hist(rollArray,bins=(np.max(rollArray-np.min(rollArray))))
plt.xlabel('Dice Roll Total')
plt.ylabel('Frequency')
plt.show()
```

Create the running average

```python
total = 0
avgArray = []
for i in range(0,numRolls):
  total = total + rollArray[i]
  avgArray.append(total / (i+1))
runningAvg = np.array(avgArray)
rollNumber = np.linspace(1,numRolls,numRolls)
print(rollNumber)
print(runningAvg)
```

Create a line chart of the moving average

```python
plt.plot(rollNumber,rolls,'bo',label='Roll Total')
plt.plot(rollNumber,runningAvg,'m-',label='Running Average')
plt.xlabel('Roll #')
plt.ylabel('Dice Sum')
plt.grid(True)
plt.legend()
plt.show()
```

Or Subplots

`plt.subplot(211)` - 2 high, 1 wide, first subplot

`plt.subplot(212)` - 2 high, 1 wide, second subplot

```python
plt.figure()
plt.subplot(211)
plt.plot(rollNumber,rolls,'bo',label='Roll Total')
plt.xlabel('Roll #')
plt.ylabel('Dice Sum')
plt.grid(True)
plt.legend()

plt.subplot(212)
plt.plot(rollNumber,runningAvg,'m-',label='Running Average')
plt.xlabel('Roll #')
plt.ylabel('Dice Sum Average')
plt.grid(True)
plt.legend()
plt.show()
```
